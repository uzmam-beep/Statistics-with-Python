```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv("phone_battery_life.txt")

# Calculate statistics for Amy's phone
amy_battery_life = data['Amy']
amy_mean = np.mean(amy_battery_life)
amy_median = np.median(amy_battery_life)
amy_range = np.max(amy_battery_life) - np.min(amy_battery_life)
amy_std_dev = np.std(amy_battery_life)

# Calculate statistics for Brian's phone
brian_battery_life = data['Brian']
brian_mean = np.mean(brian_battery_life)
brian_median = np.median(brian_battery_life)
brian_range = np.max(brian_battery_life) - np.min(brian_battery_life)
brian_std_dev = np.std(brian_battery_life)

# Print the results
print("Amy's phone battery life:")
print("Mean:", amy_mean)
print("Median:", amy_median)
print("Range:", amy_range)
print("Standard Deviation:", amy_std_dev)
print("\nBrian's phone battery life:")
print("Mean:", brian_mean)
print("Median:", brian_median)
print("Range:", brian_range)
print("Standard Deviation:", brian_std_dev)

# Create histogram
plt.hist(amy_battery_life, alpha=0.5, label='Amy')
plt.hist(brian_battery_life, alpha=0.5, label='Brian')
plt.xlabel('Battery Life')
plt.ylabel('Frequency')
plt.title('Distribution of Phone Battery Life')
plt.legend()
plt.show()
```
